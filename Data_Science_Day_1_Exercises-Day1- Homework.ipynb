```python
import csv
```

```python
def open_csv(filename, d=','):
    data=[]
    with open(filename) as whitesox_data:
        whitesox_reader = csv.reader(whitesox_data, delimiter=d)
        
        for row in whitesox_reader:
            data.append(row)
    return data 
```
